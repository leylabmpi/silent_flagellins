# Load

````r
# flagellin seqs
flg_seq = Fread(flg_seq_file, header=FALSE) %>%
    filter.(grepl('^>', V1)) %>%
    mutate.(V1 = gsub('^>', '', V1))
flg_seq
````

````r
# flagellin binding/activity
ba = Fread(ba_file) %>%
    mutate.(flg_status = case_when.(avg.activity > -2.066 ~ 'Active',
                                    avg.activity <= -2.066 & avg.binding <= 0.1144 ~ 'Evader',
                                    avg.activity <= -2.066 & avg.binding > 0.1114 ~ 'Silent'))
ba$flg_status %>% table
ba
````

````r
# flagellin GTDB-r202 taxonomy
flg_tax = Fread(flg_tax_file) %>%
    select.(-V2, -V5) %>%
    rename.('query' = V1,
            'e_value' = V3,
            'lineage' = V4) %>%
    separate.(lineage, taxonomy_levels(), sep=';') %>%
    mutate.(Phylum = ifelse(Phylum == '', 'Unclassified_bacteria', Phylum)) %>% 
    mutate.(across.(.cols = c(Phylum, Class, Order, Family, Genus, Species), 
                    ~ ifelse(.x == '' | is.na(.x), 'Unclassified', .x))) 
flg_tax
````

````r
# sample metadata
meta = Fread(meta_file) %>%
    mutate.(Sample = gsub('[.+-]+', '_', Sample)) %>%
    filter.(!is.na(non_westernized)) %>%
    mutate.(westernized = non_westernized == 'no') %>%
    select.(-non_westernized)
meta
````

### Activity/binding

#### activity & binding gradient

````r
# summarizing
hits_s = hits %>%
    summarize.(RPKM = median(RPKM),
               .by = c(sseqid, avg.activity, avg.binding)) %>%
    pivot_longer.(cols=c(avg.activity, avg.binding)) %>%
    filter.(!is.na(value))
hits_s
````

````r
# plotting
p = hits_s %>%
    ggplot(aes(RPKM, value)) +
    geom_point(alpha=0.5) +
    facet_wrap(~ name, scales='free_y') +
    theme_bw()
p.dims(6.5,3)
plot(p)
````

````r
# stats
LM = function(dt){
    # lm 
    res = lm(value ~ RPKM, data=dt) %>% summary
    df = as.data.frame(res$coefficients)
    colnames(df) = c('estimate', 'std_error', 't_value', 'p_value')
    df$coefficient = rownames(df)
    df$name = dt$name[1]
    df$adj_Rsq = res$adj.r.squared
    # spearman
    res = cor.test(dt$value, dt$RPKM, method='spearman') 
    df$spearman_pval = res$p.value
    return(df)
               
}
lm_res = hits_s %>%
    group_split.(name) %>%
    lapply(LM) %>%
    rbindlist(use.names=TRUE)
df.dims(20)
lm_res
df.dims()
````

````r
# stats
LM = function(dt){
    # lm 
    res = lm(value ~ prev, data=dt) %>% summary
    df = as.data.frame(res$coefficients)
    colnames(df) = c('estimate', 'std_error', 't_value', 'p_value')
    df$coefficient = rownames(df)
    df$name = dt$name[1]
    df$adj_Rsq = res$adj.r.squared
    # spearman
    res = cor.test(dt$value, dt$prev, method='spearman') 
    df$spearman_pval = res$p.value
    return(df)         
}

lm_res = hits_s %>%
    group_split.(name) %>%
    lapply(LM) %>%
    rbindlist(use.names=TRUE)
df.dims(20)
lm_res
df.dims()
````
